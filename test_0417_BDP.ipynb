## 4차시 (정리기에 mpg 데이터 좀 있음)

```python
import pandas as pd
import numpy as np
df = pd.DataFrame([[np.nan,2,np.nan,0],
                   [3,4,np.nan,1],
                   [np.nan, np.nan, np.nan, np.nan],
                   [np.nan, 3, np.nan, 4]],
                   columns=list('ABCD'))
df
```

```python
df.fillna(10)
```

```python
df.fillna('missing', limit=1,axis=1)
# limit = 각 열 별로 최대 1개의 결측값만 채울 수 있게 + axis=1이라 행별로 결측값 1개인듯
```

fillna(): 결측값을 채우는 함수입니다.
'missing': 결측값을 대체할 값으로, 여기서는 문자열 'missing'입니다.
limit=1: 각 열(column) 별로 최대 1개의 결측값만 대체합니다.
axis=1: axis 매개변수는 연산을 수행할 축을 지정합니다. axis=1은 열(column)을 의미합니다. 따라서, 이 코드는 각 열 별로 결측값을 채웁니다.
